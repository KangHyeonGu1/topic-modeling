```python
stop.extend(["잘","우리","했다","해서","있다","한다","생","됩니다","번","때","이러한","하는","입니다","있습니다","지능","인공","인공 지능"])
```

```python
import konlpy
import re
def tokenize_korean_text(text):
    text = re.sub(r'[^,.?!\w\s]','', text)

    okt = konlpy.tag.Okt()
    Okt_morphs = okt.pos(text)

    words = []
    for word, pos in Okt_morphs:
        if  pos == 'Noun' and  (len(word) > 1 and word not in stop):
            words.append(word)
    words_str = ' '.join(words)
    ## word를 이어붙인 string 형태가 아닌 word의 list를 return해주는 게 sklearn lda 준비 과정과의 차이

    return words_str


tokenized_list = []

for text in df1['text']:
    tokenized_list.append(tokenize_korean_text(text))

print(len(tokenized_list))
print(tokenized_list[1])
```

```python
import json
with open('tokenized_list.json', 'w') as f:
    json.dump(tokenized_list, f)
```

```python
tokenized_list2 = [sublist for sublist in tokenized_list if '자비스' not in sublist]
```
